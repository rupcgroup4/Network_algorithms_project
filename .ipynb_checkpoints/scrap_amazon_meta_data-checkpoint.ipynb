```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(columns=['title', 'group'])
```

```python
d = {'title': False, 'group': False}
id = False

with open("./amazon/amazon-meta.txt") as f:
    for line in f:
        if 'Id' in line:
            try:
                id = line.split(':')[1]
            except:
                print(line)
                break
        if 'title' in line:
            try:
                d['title'] = line.split(':')[1]
            except:
                print(line)
                break
        if 'group' in line:
            try:
                d['group'] = line.split(':')[1]
            except:
                print(line)
                break
        if id and d['title'] and d['group']:
            df.at[id, :] = d
            break
        else:
            print(f"{id}, {d['title']}, {d['group']}")
        
df.head()
        
```
