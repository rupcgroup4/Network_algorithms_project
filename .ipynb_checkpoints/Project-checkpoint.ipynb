```python
orders = orders_df.groupby('order_id')['product_id'].unique()
```

```python
prodcut_edges = []

for order in orders:
    if len(order) > 1: 
        for i in range(len(order)):
            for j in range(i+1, len(order)):
                if order[i] != order[j]:
                    prodcut_edges.append([order[i], order[j]])
            
print(len(prodcut_edges))
```

```python
with open('test.csv', 'w') as f:
    f.write(f'node1,node2')
    f.write('\n')
    for egde in prodcut_edges:
        f.write(f'{egde[0]},{egde[1]}')
        f.write('\n')
```

```python
Gcc = sorted(nx.connected_components(g), key=len, reverse=True)
print(len(Gcc[5]))
```

```python
nx.betweenness_centrality(g.subgraph(Gcc[0]))
```

```python
communities_generator = nx.community.girvan_newman(g.subgraph(Gcc[0]))
top_level_communities = next(communities_generator)
next_level_communities = next(communities_generator)
comm = sorted(map(sorted, next_level_communities),key=len, reverse=True)
```

```python
nx.community.modularity(g.subgraph(Gcc[0]), comm)
```
